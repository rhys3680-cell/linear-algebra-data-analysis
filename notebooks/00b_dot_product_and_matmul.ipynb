```python
import numpy as np

A = np.arange(12).reshape(3,4)
print(A)
print(A[0])
print(A[1])
```

```python
print(np.dot(A[0], A[1]))
print(np.dot(A[0], A[2]))
print(np.dot(A[1], A[0]))
print(np.dot(A[1], A[2]))
```

```python
print((A[0]*A[1]).sum())
print(A[0]*A[1])
```

```python
print(A)
print(A.T)

print((A@A.T)[0,0])
print((A@A.T)[0,1])
print((A@A.T)[1,0])
print((A@A.T)[1,1])
print((A@A.T)[0,2])
print((A@A.T)[2,0])
print((A@A.T)[1,2])
print((A@A.T)[2,1])
print((A@A.T)[2,2])
```

```python
G = np.zeros((3, 3))
print(G)

for i in range(3):
    for j in range(3):
        G[i, j] = np.dot(A[i], A[j])

print(G)
```

```python
print(np.diag(A@A.T))
print(np.diagonal(A@A.T))
print(np.linalg.norm(A[0]))
print(np.linalg.norm(A[1]))
print(np.linalg.norm(A[2]))

print(np.linalg.norm(A, axis=1)**2)
```

```python
print((A@A.T)[1,2])
print(np.dot(A[1],A[2]))
print(A[1] * A[2])
print((A[1] * A[2]).sum())
```

```python
total = 0
for k in range(4):
    total += A[1][k] * A[2][k]   # 한 자리씩 곱해서 누적
    print(f"{A[1][k]} * {A[2][k]} = {A[1][k]*A[2][k]}, 누적합 = {total}")
print("최종:", total)
```

```python
print(A[1] @ A[2])
```

```python
print(A@A.T)
```

```python
print(np.allclose(G, G.T))
```

```python
print(A / np.linalg.norm(A, axis=1, keepdims=True))
A_unit = A / np.linalg.norm(A, axis=1, keepdims=True)
print(A_unit)
print(A_unit @ A_unit.T)
```
