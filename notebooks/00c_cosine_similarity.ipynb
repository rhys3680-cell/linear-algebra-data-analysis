```python
import numpy as np

A = np.arange(12).reshape(3, 4)

print(A)

print(np.linalg.norm(A, axis=1, keepdims=True))

A_unit = A / np.linalg.norm(A, axis=1, keepdims=True)

print(A_unit)
print(np.linalg.norm(A_unit, axis=1))
```

```python
print(A_unit @ A_unit.T)
print(np.diag(A_unit @ A_unit.T))
```
