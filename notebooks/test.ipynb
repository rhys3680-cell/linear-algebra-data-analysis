## 연습장


```python
import numpy as np

A = np.arange(12).reshape(3,4)
print(A)
print(A.T)
print(A[0])
print(A.T[0])

print(A@A.T)

G = np.zeros((3,3))

print(G)

for i in range(3):
    for j in range(3):
        print(A[i])
        print(A[j])
        print(A[i] @ A[j])
        G[i, j] = A[i] @ A[j]


print(G)
```
